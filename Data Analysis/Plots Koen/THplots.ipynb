```python
# importing libraries

import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
import numpy as np
```

```python
# reading first csv and creating lists
# for first 5 graphs, with indivudual plant / AT

regelnummer = 0

# wavelength lists
wv_abslist = []
wv_flRTlist = []
wv_flLTlist = []

# absorption lists
absorptionAFlist = []
absorptionATlist = []

# fluorescence lists
fluorescenceAFRTlist = []
fluorescenceATRTlist = []
fluorescenceAFLTlist = []
fluorescenceATLTlist = []


with open('ValuesSimplified.csv', 'r') as input:
    for regel in input:
        if regelnummer > 0:

            # absorption lists
            data_opgeknipt = regel.split(';')
            waves = data_opgeknipt[0].replace(',', '.')
            wv_abs = float(waves)
            abs = data_opgeknipt[1].replace(',', '.')
            absorptionAF = float(abs)
            abs2 = data_opgeknipt[2].replace(',', '.')
            absorptionAT = float(abs2)

            wv_abslist.append(wv_abs)
            absorptionAFlist.append(absorptionAF)
            absorptionATlist.append(absorptionAT)

            # RT fluorescence lists
            if regelnummer < 202:
                waves2 = data_opgeknipt[6].replace(',', '.')
                wv_flRT = float(waves2)
                flrt = data_opgeknipt[7].replace(',', '.')
                fluorescenceAFRT = float(flrt)
                flrt2 = data_opgeknipt[8].replace(',', '.')
                fluorescenceATRT = float(flrt2)

                wv_flRTlist.append(wv_flRT)
                fluorescenceAFRTlist.append(fluorescenceAFRT)
                fluorescenceATRTlist.append(fluorescenceATRT)
            
            # LT fluorescence lists
            if regelnummer < 252:
                waves3 = data_opgeknipt[3].replace(',', '.')
                wv_flLT = float(waves3)
                fllt = data_opgeknipt[4].replace(',', '.')
                fluorescenceAFLT = float(fllt)
                fllt2 = data_opgeknipt[5].replace(',', '.')
                fluorescenceATLT = float(fllt2)

                wv_flLTlist.append(wv_flLT)
                fluorescenceAFLTlist.append(fluorescenceAFLT)
                fluorescenceATLTlist.append(fluorescenceATLT)
        
        regelnummer += 1

```

```python
# reading second csv and creating lists
# for last three graphs, with all plants

regelnummer2 = 0

# absorption lists
ab_Elist = []
ab_Jlist = []
ab_Mlist = []
ab_Klist = []
ab_Nlist = []
ab_Tlist = []

# RT fluorescence lists
flRT_FAlist = []
flRT_ATlist = []
flRT_AFlist = []
flRT_CElist = []
flRT_HAlist = []
flRT_HPlist = []

# LT fluorescence lists
flLT_AFlist = []
flLT_CElist = []
flLT_HAlist = []
flLT_HPlist = []
flLT_FAlist = []
flLT_ATlist = []

with open('AllPlantsData.csv', 'r') as input:
    for regel in input:
        if regelnummer2 > 0:
            
            # absorption lists
            data_opgeknipt = regel.split(';')
            ab1 = data_opgeknipt[0].replace(',', '.')
            ab_E = float(ab1)
            ab2 = data_opgeknipt[1].replace(',', '.')
            ab_J = float(ab2)
            ab3 = data_opgeknipt[2].replace(',', '.')
            ab_M = float(ab3)
            ab4 = data_opgeknipt[3].replace(',', '.')
            ab_K = float(ab4)
            ab5 = data_opgeknipt[4].replace(',', '.')
            ab_N = float(ab5)
            ab6 = data_opgeknipt[5].replace(',', '.')
            ab_T = float(ab6)

            ab_Elist.append(ab_E)
            ab_Jlist.append(ab_J)
            ab_Mlist.append(ab_M)
            ab_Klist.append(ab_K)
            ab_Nlist.append(ab_N)
            ab_Tlist.append(ab_T)

            # 4 RT fluorescence lists
            if regelnummer2 < 202:
                fr1 = data_opgeknipt[14].replace(',', '.')
                flRT_AF = float(fr1)
                fr2 = data_opgeknipt[15].replace(',', '.')
                flRT_CE = float(fr2)
                fr3 = data_opgeknipt[16].replace(',', '.')
                flRT_HA = float(fr3)
                fr4 = data_opgeknipt[17].replace(',', '.')
                flRT_HP = float(fr4)

                flRT_AFlist.append(flRT_AF)
                flRT_CElist.append(flRT_CE)
                flRT_HAlist.append(flRT_HA)
                flRT_HPlist.append(flRT_HP)
            
            # 2 RT fluorescence lists + LT fluorescence lists
            if regelnummer2 < 252:
                fr5 = data_opgeknipt[12].replace(',', '.')
                flRT_FA = float(fr5)
                fr6 = data_opgeknipt[13].replace(',', '.')
                flRT_AT = float(fr6)
                fl1 = data_opgeknipt[6].replace(',', '.')
                flLT_AF = float(fl1)
                fl2 = data_opgeknipt[7].replace(',', '.')
                flLT_CE = float(fl2)
                fl3 = data_opgeknipt[8].replace(',', '.')
                flLT_HA = float(fl3)
                fl4 = data_opgeknipt[9].replace(',', '.')
                flLT_HP = float(fl4)
                fl5 = data_opgeknipt[10].replace(',', '.')
                flLT_FA = float(fl5)
                fl6 = data_opgeknipt[11].replace(',', '.')
                flLT_AT = float(fl6)

                flRT_FAlist.append(flRT_FA)
                flRT_ATlist.append(flRT_AT)
                flLT_AFlist.append(flLT_AF)
                flLT_CElist.append(flLT_CE)
                flLT_HAlist.append(flLT_HA)
                flLT_HPlist.append(flLT_HP)
                flLT_FAlist.append(flLT_FA)
                flLT_ATlist.append(flLT_AT)

        regelnummer2 += 1
```

```python
# reading third csv and creating lists
# for PSI graphs

regelnummer3 = 0

# absorption lists
PSI_abElist = []
PSI_abJlist = []
PSI_abMlist = []
PSI_abKlist = []
PSI_abNlist = []
PSI_abTlist = []
PSI_abClist = []

# fluorescence lists
PSI_flElist = []
PSI_flJlist = []
PSI_flMlist = []
PSI_flKlist = []
PSI_flNlist = []
PSI_flTlist = []

with open('Data PSI 2.csv', 'r') as input:
    for regel in input:
        if regelnummer3 > 0:

            # absorption lists
            data_opgeknipt = regel.split(';')
            PSIab1 = data_opgeknipt[0].replace(',', '.')
            PSIabN = float(PSIab1)
            PSIab2 = data_opgeknipt[1].replace(',', '.')
            PSIabT = float(PSIab2)

            PSI_abNlist.append(PSIabN)
            PSI_abTlist.append(PSIabT)


            if regelnummer3 < 1002:
                PSIab3 = data_opgeknipt[2].replace(',', '.')
                PSIabE = float(PSIab3)
                PSIab4 = data_opgeknipt[3].replace(',', '.')
                PSIabJ = float(PSIab4)
                PSIab5 = data_opgeknipt[4].replace(',', '.')
                PSIabM = float(PSIab5)
                PSIab6 = data_opgeknipt[5].replace(',', '.')
                PSIabK = float(PSIab6)
                PSI_AC = data_opgeknipt[6].replace(',', '.')
                PSIabC = float(PSI_AC)
                
                PSI_abElist.append(PSIabE)
                PSI_abJlist.append(PSIabJ)
                PSI_abMlist.append(PSIabM)
                PSI_abKlist.append(PSIabK)
                PSI_abClist.append(PSIabC)

            # LT fluorescence lists
            if regelnummer3 < 202:
                PSIfl1 = data_opgeknipt[8].replace(',', '.')
                PSIflK = float(PSIfl1)

                PSI_flKlist.append(PSIflK)

            if regelnummer3 < 252:
                PSIfl2 = data_opgeknipt[7].replace(',', '.')
                PSIflT = float(PSIfl2)

                PSI_flTlist.append(PSIflT)

        
        regelnummer3 += 1

ShortList = wv_abslist[125:]
```

```python
# converting lists to numpy arrays

# absorption and fluorescence lists for first 5 graphs
absAF_final = np.array(absorptionAFlist)
absAT_final = np.array(absorptionATlist)
flAFRT_final = np.array(fluorescenceAFRTlist)
flAFLT_final = np.array(fluorescenceAFLTlist)

# absorption lists from all plants
ab_E_final = np.array(ab_Elist)
ab_J_final = np.array(ab_Jlist)
ab_M_final = np.array(ab_Mlist)
ab_K_final = np.array(ab_Klist)
ab_N_final = np.array(ab_Nlist)
ab_T_final = np.array(ab_Tlist)

# RT fluorescence lists from all plants
flRT_FA_final = np.array(flRT_FAlist)
flRT_AT_final = np.array(flRT_ATlist)
flRT_AF_final = np.array(flRT_AFlist)
flRT_CE_final = np.array(flRT_CElist)
flRT_HA_final = np.array(flRT_HAlist)
flRT_HP_final = np.array(flRT_HPlist)

# LT fluorescence lists from all plants
flLT_FA_final = np.array(flLT_FAlist)
flLT_AT_final = np.array(flLT_ATlist)
flLT_AF_final = np.array(flLT_AFlist)
flLT_CE_final = np.array(flLT_CElist)
flLT_HA_final = np.array(flLT_HAlist)
flLT_HP_final = np.array(flLT_HPlist)

# PSI absorption lists from all plants
PSI_abEfinal = np.array(PSI_abElist)
PSI_abJfinal = np.array(PSI_abJlist)
PSI_abMfinal = np.array(PSI_abMlist)
PSI_abKfinal = np.array(PSI_abKlist)
PSI_abNfinal = np.array(PSI_abNlist)
PSI_abTfinal = np.array(PSI_abTlist)
PSI_abCfinal = np.array(PSI_abClist)

# PSI fluorescence lists (AF and AT)
PSI_flKfinal = np.array(PSI_flKlist)
PSI_flTfinal = np.array(PSI_flTlist)
```

```python
# creating normalization functions for all arrays

# normalization functions for first 5 graphs
# absorption AF / AT and fluorescence AF at certain peaks
norm = Normalize()
ab_normAF = Normalize(vmax = 0.381760925, clip=False)
ab_normAT = Normalize(vmax = 0.236148804, clip=False)
normAFLT = Normalize(vmax = 8157200.5)
normAFLT2 = Normalize(vmax = 6138380, clip=False)

# normalization functions for all absorptions
abnorm_E = Normalize(vmax = 0.725952387, clip=False)
abnorm_J = Normalize(vmax = 0.499314457, clip=False)
abnorm_M = Normalize(vmax = 0.537089109, clip=False)
abnorm_K = Normalize(vmax = 0.381760925, clip=False)
abnorm_N = Normalize(vmax = 0.215625808, clip=False)
abnorm_T = Normalize(vmax = 0.236148804, clip=False)

# normalizing to the maximum, used for RT fluorescence
# (you need a new one for every array)
norm1 = Normalize()
norm2 = Normalize()
norm3 = Normalize()
norm4 = Normalize()
norm5 = Normalize()
norm6 = Normalize()

# normalization functions for all LT fluorescence
LTnorm_FA = Normalize(vmax = 49243912, clip=False)
LTnorm_AT = Normalize(vmax = 1, clip=False)
LTnorm_AF = Normalize(vmax = 8157200.5, clip=False)
LTnorm_CE = Normalize(vmax = 37699076, clip=False)
LTnorm_HA = Normalize(vmax = 11919807, clip=False)
LTnorm_HP = Normalize(vmax = 47526864, clip=False)

# normalization functions for all PSI absorption
PSInormE = Normalize(vmax = 0.581648707, clip=False)
PSInormJ = Normalize(vmax = 0.364870429, clip=False)
PSInormM = Normalize(vmax = 1.506798625, clip=False)
PSInormK = Normalize(vmax = 0.777778506, clip=False)
PSInormN = Normalize(vmax = 0.6563747911, clip=False)
PSInormT = Normalize(vmax = 0.7831941247, clip=False)
PSInormC = Normalize(vmax = 1.343286753, clip=False)

# normalization functions for PSI LT fluorescence (AF and AT)
PSInorm_flK = Normalize(vmax = 14988751, clip=False)
```

```python
# normalizing the arrays with the given normalization functions

# normalization for first 5 graphs
absAFnorm = ab_normAF(absAF_final)
absATnorm = ab_normAT(absAT_final)
flAFRTnorm = norm(flAFRT_final)
flAFLTnorm = normAFLT(flAFLT_final)
flAFLTnorm2 = normAFLT2(flAFLT_final)

# normalization of absorption arrays
ab_Enorm = abnorm_E(ab_E_final)
ab_Jnorm = abnorm_J(ab_J_final)
ab_Mnorm = abnorm_M(ab_M_final)
ab_Knorm = abnorm_K(ab_K_final)
ab_Nnorm = abnorm_N(ab_N_final)
ab_Tnorm = abnorm_T(ab_T_final)

# normalization of RT fluorescence arrays
fl_RTFAnorm = norm1(flRT_FA_final)
fl_RTATnorm = norm2(flRT_AT_final)
fl_RTAFnorm = norm3(flRT_AF_final)
fl_RTCEnorm = norm4(flRT_CE_final)
fl_RTHAnorm = norm5(flRT_HA_final)
fl_RTHPnorm = norm6(flRT_HP_final)

# normalization of LT fluorescence arrays
fl_LTFAnorm = LTnorm_FA(flLT_FA_final)
fl_LTATnorm = LTnorm_AT(flLT_AT_final)
fl_LTAFnorm = LTnorm_AF(flLT_AF_final)
fl_LTCEnorm = LTnorm_CE(flLT_CE_final)
fl_LTHAnorm = LTnorm_HA(flLT_HA_final)
fl_LTHPnorm = LTnorm_HP(flLT_HP_final)

# normalization of PSI absorption arrays
PSI_Enorm = PSInormE(PSI_abEfinal)
PSI_Jnorm = PSInormJ(PSI_abJfinal)
PSI_Mnorm = PSInormM(PSI_abMfinal)
PSI_Knorm = PSInormK(PSI_abKfinal)
PSI_Nnorm = PSInormN(PSI_abNfinal)
PSI_Tnorm = PSInormT(PSI_abTfinal)
PSI_Cnorm = PSInormC(PSI_abCfinal)

# normalization of PSI fluorescence arrays
PSI_flKnorm = PSInorm_flK(PSI_flKfinal)
```

```python
# plotting first graph

# absorption vs RT fluorescence of AF
plt.figure(1)
plt.plot(wv_abslist, absAFnorm, 'b-')
plt.plot(wv_flRTlist, flAFRTnorm, 'r-')
plt.xlabel('wavelength (nm)')
plt.ylabel('absorption / fluorescence (a.u)')
plt.xlim(350, 800)
plt.title('Absorption vs RT fluorescence of AF')
plt.legend(['absorption', 'fluorescence (RT)'])
plt.savefig('absorption vs RT fluorescence')
```

```python
# plotting second graph

# calculating difference between AT and AF
diffAT_AF_abslist = absATnorm - absAFnorm

# absorption of AF vs AT (including difference)
plt.figure(2)
plt.plot(wv_abslist, absAFnorm, 'b-')
plt.plot(wv_abslist, absATnorm, 'r-' )
plt.plot(wv_abslist, diffAT_AF_abslist, 'k-')
plt.xlabel('wavelength (nm)')
plt.ylabel('absorption (a.u)')
plt.xlim(350, 800)
plt.title('Absorption of AF vs AT')
plt.legend(['absorption AF', 'absorption AT', 'difference (AT - AF)'])
plt.savefig('absorption AF vs AT')
```

```python
# plotting third graph

# calculating difference between AT and AF
diffAT_AF_fllist = fluorescenceATRTlist - flAFRTnorm

# fluorescence of AF vs RT (RT, including difference)
plt.figure(3)
plt.plot(wv_flRTlist, flAFRTnorm, 'b-')
plt.plot(wv_flRTlist, fluorescenceATRTlist, 'r-')
plt.plot(wv_flRTlist, diffAT_AF_fllist, 'k-')
plt.xlabel('wavelength (nm)')
plt.ylabel('fluorescence (a.u)')
plt.xlim(602.3083, 802.6123)
plt.title('Fluorescence of AF vs AT (RT)')
plt.legend(['fluorescence AF', 'fluorescence AT', 'difference (AT - AF)'])
plt.savefig('fluorescenceRT AFvAT')
```

```python
# plotting fourth graph

# calculating difference between AT and AF
diffLT_AT_AF_fllist = fluorescenceATLTlist - flAFLTnorm

# fluorescence of AF vs AT (LT, including difference)
plt.figure(4)
plt.plot(wv_flLTlist, flAFLTnorm, 'b-')
plt.plot(wv_flLTlist, fluorescenceATLTlist, 'r-')
plt.plot(wv_flLTlist, diffLT_AT_AF_fllist, 'k-')
plt.xlabel('wavelength (nm)')
plt.ylabel('fluorescence (a.u)')
plt.xlim(602.3083, 852.6883)
plt.title('Fluorescence of AF vs AT (LT)')
plt.legend(['fluorescence AF', 'fluorescence AT', 'difference (AT - AF)'])
plt.savefig('fluorescenceLT AFvAT')
```
